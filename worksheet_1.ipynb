**Problem - 1: Array Creation:**

Complete the following Tasks:
1.	Initialize an empty array with size 2X2.
2.	Initialize an all one array with size 4X2.
3.	Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4.	Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}

6.	For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}



```python
import numpy as np

# 1. Initialize an empty array with size 2x2
empty_array = np.empty((2, 2))
print("Empty array (2x2):")
print(empty_array)

# 2. Initialize an all-one array with size 4x2
ones_array = np.ones((4, 2))
print("\nAll-ones array (4x2):")
print(ones_array)

# 3. Return a new array of given shape and type, filled with fill value (using np.full)
shape = (3, 3)
fill_value = 7
full_array = np.full(shape, fill_value)
print("\nArray filled with value 7 (3x3):")
print(full_array)

# 4. Return a new array of zeros with same shape and type as a given array (using np.zeros_like)
reference_array = np.array([[1, 2, 3], [4, 5, 6]])
zeros_like_array = np.zeros_like(reference_array)
print("\nZeros array with the same shape as reference array:")
print(zeros_like_array)

# 5. Return a new array of ones with same shape and type as a given array (using np.ones_like)
ones_like_array = np.ones_like(reference_array)
print("\nOnes array with the same shape as reference array:")
print(ones_like_array)

# 6. Convert an existing list to a numpy array
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\nNumpy array from list [1, 2, 3, 4]:")
print(numpy_array)
```

For the following arrays:

{Hint:try np.concatenate() or np.vstack() functions.

```python
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
array_sum = x + y
print("Sum of x and y:")
print(array_sum)

# 2. Subtract the two arrays
array_diff = x - y
print("\nDifference of x and y:")
print(array_diff)

# 3. Multiply the array x with any integer (choosing 3)
array_multiplied = x * 3
print("\nArray x multiplied by 3:")
print(array_multiplied)

# 4. Find the square of each element of the array x
array_squared = x ** 2
print("\nSquare of each element in x:")
print(array_squared)

# 5. Find the dot product
dot_v_w = np.dot(v, w)  # Dot product between v and w
dot_x_v = np.dot(x, v)  # Dot product between x and v
dot_x_y = np.dot(x, y)  # Dot product between x and y
print("\nDot product of v and w:", dot_v_w)
print("Dot product of x and v:")
print(dot_x_v)
print("Dot product of x and y:")
print(dot_x_y)

# 6. Concatenate x and y along rows and v and w along columns
concat_x_y_row = np.concatenate((x, y), axis=0)
concat_v_w_column = np.vstack((v, w))  # Or np.column_stack((v, w))
print("\nConcatenation of x and y along rows:")
print(concat_x_y_row)
print("Concatenation of v and w along columns:")
print(concat_v_w_column)

# 7. Concatenate x and v
try:
    concat_x_v = np.concatenate((x, v), axis=0)
    print("\nConcatenation of x and v along rows:")
    print(concat_x_v)
except ValueError as e:
    print("\nError in concatenating x and v:")
    print(e)
    print("Explanation: x has shape (2, 2), and v has shape (2,). "
          "Concatenation requires matching dimensions, and here, "
          "v must be reshaped to (1, 2) to match x along the row axis.")
```

A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]); Prove following with Numpy:

3.	Prove (AB)T = BTAT.

•	Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1 x − y + 2z = −3 3x + y − z = 9
{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

•	Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function of Numpy}


```python
import numpy as np

# Given arrays
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A⁻¹ = I
A_inv = np.linalg.inv(A)
result_1 = np.dot(A, A_inv)
print("Prove A * A⁻¹ = I:")
print(result_1)  # Should be the identity matrix

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nProve AB ≠ BA:")
print("AB:")
print(AB)
print("BA:")
print(BA)
print("Are AB and BA equal?", np.array_equal(AB, BA))

# 3. Prove (AB)^T = B^T * A^T
AB_T = np.transpose(AB)
B_T_A_T = np.dot(B.T, A.T)
print("\nProve (AB)^T = B^T * A^T:")
print("(AB)^T:")
print(AB_T)
print("B^T * A^T:")
print(B_T_A_T)
print("Are (AB)^T and B^T * A^T equal?", np.array_equal(AB_T, B_T_A_T))
```

```python
# Define the coefficient matrix A and the constants matrix B
A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B = np.array([-1, -3, 9])

# Solve using np.linalg.inv
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("\nSolution using np.linalg.inv:")
print("X = [x, y, z]:", X)

# Verify solution using np.linalg.solve
X_solve = np.linalg.solve(A, B)
print("\nSolution using np.linalg.solve:")
print("X = [x, y, z]:", X_solve)

# Verify that both methods give the same result
print("\nAre the two solutions equal?", np.allclose(X, X_solve))
```

4.2	Experiment: How Fast is Numpy?
In this exercise, you will compare the performance and implementation of operations using plain Python lists (arrays) and NumPy arrays. Follow the instructions:
1. Element-wise Addition:
•	Using Python Lists, perform element-wise addition of two lists of size 1,000,000. Measure and Print the time taken for this operation.
•	Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.
2. Element-wise Multiplication
•	Using Python Lists, perform element-wise multiplication of two lists of size 1,000,000. Measure and Print the time taken for this operation.
•	Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.
3. Dot Product
•	Using Python Lists, compute the dot product of two lists of size 1,000,000. Measure and Print the time taken for this operation.
•	Using Numpy Arrays, Repeat the calculation and measure and print the time taken for this operation.
4. Matrix Multiplication
•	Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.
•	Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.


```python
import numpy as np
import time

# Helper function to measure execution time
def measure_time(func, *args, **kwargs):
    start = time.time()
    result = func(*args, **kwargs)
    end = time.time()
    return result, end - start

# 1. Element-wise Addition
size = 1_000_000

# Using Python Lists
list1 = list(range(size))
list2 = list(range(size))
_, time_list_add = measure_time(lambda: [list1[i] + list2[i] for i in range(size)])

# Using NumPy Arrays
array1 = np.arange(size)
array2 = np.arange(size)
_, time_numpy_add = measure_time(lambda: array1 + array2)

print(f"Element-wise addition time (Python lists): {time_list_add:.4f} seconds")
print(f"Element-wise addition time (NumPy arrays): {time_numpy_add:.4f} seconds")

# 2. Element-wise Multiplication
# Using Python Lists
_, time_list_mult = measure_time(lambda: [list1[i] * list2[i] for i in range(size)])

# Using NumPy Arrays
_, time_numpy_mult = measure_time(lambda: array1 * array2)

print(f"\nElement-wise multiplication time (Python lists): {time_list_mult:.4f} seconds")
print(f"Element-wise multiplication time (NumPy arrays): {time_numpy_mult:.4f} seconds")

# 3. Dot Product
# Using Python Lists
_, time_list_dot = measure_time(lambda: sum(list1[i] * list2[i] for i in range(size)))

# Using NumPy Arrays
_, time_numpy_dot = measure_time(lambda: np.dot(array1, array2))

print(f"\nDot product time (Python lists): {time_list_dot:.4f} seconds")
print(f"Dot product time (NumPy arrays): {time_numpy_dot:.4f} seconds")

# 4. Matrix Multiplication
matrix_size = 1000

# Using Python Lists
matrix1 = [[j for j in range(matrix_size)] for i in range(matrix_size)]
matrix2 = [[j for j in range(matrix_size)] for i in range(matrix_size)]
_, time_list_matrix_mult = measure_time(
    lambda: [[sum(matrix1[i][k] * matrix2[k][j] for k in range(matrix_size)) for j in range(matrix_size)] for i in range(matrix_size)]
)

# Using NumPy Arrays
np_matrix1 = np.arange(matrix_size ** 2).reshape(matrix_size, matrix_size)
np_matrix2 = np.arange(matrix_size ** 2).reshape(matrix_size, matrix_size)
_, time_numpy_matrix_mult = measure_time(lambda: np.dot(np_matrix1, np_matrix2))

print(f"\nMatrix multiplication time (Python lists): {time_list_matrix_mult:.4f} seconds")
print(f"Matrix multiplication time (NumPy arrays): {time_numpy_matrix_mult:.4f} seconds")
```
